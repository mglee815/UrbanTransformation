```python
import pandas as pd
import numpy as np  
import glob  
import sys
import warnings
import os
import time
```

```python
key_lst = []
feat_lst = []
f_lst = []
for f in glob.glob(f'/Users/mglee/Desktop/folder/merged_data/*.pkl'): # 예를들어 201901, 201902 로 된 파일이면 2019_*  
    df = pd.read_pickle(f)
    col = df.columns
    key = df['키워드']
    keywords = ""
    features = ""
    for item in key:
        try:
            keywords += item.replace("," , " ")
        except:
            pass
        
    feature = df['특성추출(가중치순 상위 50개)']
    for item in feature:
        try:
            features += item.replace(",", " ")
        except:
            pass
    
    key_lst.append(keywords)
    feat_lst.append(features)
    f_lst.append(f)
    
```

```python
print(f" key_lst : {len(key_lst)}, feat_lst : {len(feat_lst)}, f : {f_lst}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
T_Vec = TfidfVectorizer(min_df= 5, analyzer='word', sublinear_tf= True)
DTM = T_Vec.fit_transform(key_lst)
```

```python
tfidf = pd.DataFrame(DTM.toarray(), columns = T_Vec.get_feature_names())
```
